```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
%matplotlib inline


def perspective_transform(img):
    img_size = (img.shape[1], img.shape[0])

    src = np.float32(
        [[200, 720],
        [1100, 720],
        [595, 450],
        [685, 450]])
    
    dst = np.float32(
        [[300, 720],
        [980, 720],
        [300, 0],
        [980, 0]])

    #变换矩阵
    m = cv2.getPerspectiveTransform(src,dst)
    #m_inv = cv2.getPerspectiveTransform(dst,src)
    #根据变换矩阵进行变换
    unwarped = cv2.warpPerspective(img, m,(1280,720))
    #warped = cv2.warpPerspective(unwarped, m_inv,(1280,720))
    
    return unwarped

#将img进行投射变换，返回投射变换后的图像，投射变换逆变换后的图像，投射变换矩阵和逆矩阵
imgs = glob.glob("testImage/test*")
for per in imgs:
    dir, name = os.path.split(per)
    img = cv2.imread(per)
    unwarped = perspective_transform(img)
    cv2.imwrite("test_image_undistort_warped/" + name, unwarped)

#显示最后一个测试图像
plt.figure(figsize = (30, 30))
plt.title('perspective_transform')
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.subplot(1, 2, 2)
plt.imshow(unwarped)
plt.axis('off')
```
